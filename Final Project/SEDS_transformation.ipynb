This notebook is intended to transform the complete SEDS into something more suited for the purposes of this project, by eliminating unneeded data.

```python
import numpy as np
import pandas as pd
import regex
df = pd.DataFrame(pd.read_csv('Complete_SEDS.csv'))
df.pop('Data_Status') #delete status column, unneeded (entire column is 2020F, the document)
df = df[df.StateCode == 'WA'] #we only want data from WA, Washington state
df.head()
```

As per https://www.eia.gov/state/seds/sep_use/notes/use_guide.pdf

MSN in form AABBC

AA: Type of energy
BB: Energy Activity/Consuming Sector
C: Units/Data type
A full list of codes is found here https://www.eia.gov/state/seds/CDF/Codes_and_Descriptions.xlsx

Desired codes are:
ESACB - Electricity consumption in the transportation sector
TEACB - Total energy consumption in the transportation sector. 

From the above codes, we're only interested in

```python
df = df[(df.MSN == 'ESACB') | (df.MSN == 'TEACB')]
df.reset_index(drop=True, inplace=True)
df
```

```python
df.to_csv('SEDS_transformed.csv')
```
